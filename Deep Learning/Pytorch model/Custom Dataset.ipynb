```python
import torch
from torchvision import transforms
import numpy as np
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, X, Y, transform=None):  # X: 입력, Y: 정답
        self.X = X
        self.Y = Y
        self.transform = transform

    def __len__(self):
        return self.X.shape[0]

    def __getitem__(self, idx):
        # SGD, DataLoader에서 minibatch로 끌어낼 때 사용??
        x = self.X[idx]
        if self.transform is not None:
            x = self.transform(x)
        y = self.Y[idx]
        return x, y
```

```python
X_data = np.arange(-10, 10).reshape(-1, 1)
Y_data = X_data**2  # 레이블이 제곱, 1.1 -> 1.21
```

```python
transform = lambda x:x+1
BATCH_SIZE = 8

custom_DS = CustomDataset(X_data, Y_data, transform=transform)  # 데이터 20개 (all)
train_DS, val_DS, test_DS = torch.utils.data.random_split(custom_DS, [10, 5, 5])  # 데이터 shuffle (train: 10, val: 5, test:5)
print(len(train_DS)) # 10
print(len(val_DS))   # 5
print(len(test_DS))  # 5
print()

# 아래는 또 다른 방법 (간지)
NoT = int(len(custom_DS) * 0.8)
NoV = int(len(custom_DS) * 0.1)
NoTes = len(custom_DS) - NoT - NoV
train_DS, val_DS, test_DS = torch.utils.data.random_split(custom_DS, [NoT, NoV, NoTes])
print(len(train_DS)) # 16
print(len(val_DS))   # 2
print(len(test_DS))  # 2
```

```python
train_DL = torch.utils.data.DataLoader(train_DS, batch_size=BATCH_SIZE, shuffle=True)
val_DL = torch.utils.data.DataLoader(val_DS, batch_size=BATCH_SIZE, shuffle=True)
test_DL = torch.utils.data.DataLoader(test_DS, batch_size=BATCH_SIZE, shuffle=True)
```

```python
for x_batch, y_batch in train_DL:
    print(f"x_batch = {x_batch.reshape(-1)}, \n"
          f"y_batch = {y_batch.reshape(-1)}")
print()

for x_batch, y_batch in val_DL:
    print(f"x_batch = {x_batch.reshape(-1)}, \n"
          f"y_batch = {y_batch.reshape(-1)}")
print()

for x_batch, y_batch in test_DL:
    print(f"x_batch = {x_batch.reshape(-1)}, \n"
          f"y_batch = {y_batch.reshape(-1)}")
print()
```
