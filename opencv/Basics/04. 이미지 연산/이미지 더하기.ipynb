```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image_1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_2 = cv2.flip(image_1, 0)

fig, axes = plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(image_1)
axes[0].set_title('image_1')
axes[0].axis('off')

axes[1].imshow(image_2)
axes[1].set_title('image_2')
axes[1].axis('off')
plt.show()

"""
다른 이미지를 다운받기 귀찮아서 원본을 뒤집은 이미지 사용 
"""
```

```python
dst_1 = image_1 + image_2 # 그냥 더하는 방법 
dst_2 = cv2.add(image_1, image_2) # 함수로 더하는 방법 

fig, axes = plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(dst_1)
axes[0].set_title('image_1 + image_2')
axes[0].axis('off')

axes[1].imshow(dst_2)
axes[1].set_title('cv2.add(image_1, image_2)')
axes[1].axis('off')
plt.show()
```
