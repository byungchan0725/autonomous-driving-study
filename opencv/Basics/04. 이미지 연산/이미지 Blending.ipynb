```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image_1 = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_2 = cv2.flip(image_1, 0)

fig, axes = plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(image_1)
axes[0].set_title('image_1')
axes[0].axis('off')

axes[1].imshow(image_2)
axes[1].set_title('image_2')
axes[1].axis('off')
plt.show()

"""
  " "   
"""
```

```python
dst_1 = image_1 + image_2 #    
dst_2 = cv2.addWeighted(image_1, float(20) * 0.01, image_2, float(30) * 0.01, 0)

fig, axes = plt.subplots(2, 2, figsize=(10, 5))

for i in range(1, 3):
    for j in range(1, 3):
        dst = cv2.addWeighted(image_1, float(i*30) * 0.01, image_2, float(i*30) * 0.01, 0)
        axes[i-1, j-1].imshow(dst)
        axes[i-1, j-1].set_title(f'{(i*20, j*20)}')
        axes[i-1, j-1].axis('off')
        
plt.show()
```
