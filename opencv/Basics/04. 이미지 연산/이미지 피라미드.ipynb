```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

lower_reso = cv2.pyrDown(image) # 원본 이미지의 1/4 사이즈 
higher_reso = cv2.pyrUp(image) # 4배 

titles = ['origin', 'lower', 'higher']
images = [image, lower_reso, higher_reso]

plt.figure(figsize=(20, 10))
for i in range(len(titles)):
    plt.subplot(1, len(titles), i+1)
    plt.title(titles[i])
    plt.imshow(images[i])

plt.show()

""" 구분을 할 수 없어서 옆에 axis를 off하지 않았음 """
```
