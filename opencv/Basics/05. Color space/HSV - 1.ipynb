```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/tomato.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(image)
```

```python
titles = ['h', 's', 'v']
images = [h, s, v]

plt.figure(figsize=(20, 10))
for i in range(len(titles)):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
