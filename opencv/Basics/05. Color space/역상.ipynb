```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

dst = cv2.bitwise_not(image)

titles = ['Origin', 'dst']
images = [image, dst]

plt.figure(figsize=(10, 10))
for i in range(len(titles)):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
