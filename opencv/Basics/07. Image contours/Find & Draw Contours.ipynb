```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/contour image.jpg'
image = cv2.imread(path)
origin = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, threshold = cv2.threshold(image, 125, 255, cv2.THRESH_BINARY)

contours, hierachy = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 10)
image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

titles = ['origin', 'draw contours']
images = [origin, image]

plt.figure(figsize=(12, 10))
for i in range(len(titles)):
    plt.subplot(1, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
