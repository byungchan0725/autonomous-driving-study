```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

dst1 = cv2.flip(rgb_image, -1) # 상하좌우 대칭 
dst2 = cv2.flip(rgb_image, 1) # 상하 대칭
dst3 = cv2.flip(rgb_image, 0) # 좌우 대칭 

fig, axes = plt.subplots(2, 2, figsize=(6, 5))

axes[0, 0].imshow(rgb_image)
axes[0, 0].set_title('Origin Image')
axes[0, 0].axis('off')

axes[0, 1].imshow(dst1)
axes[0, 1].set_title('dst1 Image')
axes[0, 1].axis('off')

axes[1, 0].imshow(dst2)
axes[1, 0].set_title('dst2 Image')
axes[1, 0].axis('off')

axes[1, 1].imshow(dst3)
axes[1, 1].set_title('dst3 Image')
axes[1, 1].axis('off')

plt.show()
```
