```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

h, w, c = rgb_image.shape
print(h, w, c)
```

```python
def matrix(center, angle, scale):
    """
    center: 중심정
    angle: 돌릴 각도
    scale: 비율
    """
    return cv2.getRotationMatrix2D(center, angle, scale)

matrix_1 = matrix((h/2, w/2), 90, 1)
matrix_2 = matrix((h/3, w/5), 45, 2)

dst1 = cv2.warpAffine(rgb_image, matrix_1, (w, h))
dst2 = cv2.warpAffine(rgb_image, matrix_2, (w, h))

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(rgb_image) # origin image
axes[0].set_title('Origin Image')
axes[0].axis('off')

axes[1].imshow(dst1)
axes[1].set_title('dst1') # 스케일이 1이고, 중간을 중심점으로 90도 돌린 이미지
axes[1].axis('off')

axes[2].imshow(dst2)
axes[2].set_title('dst2') # 스케일이 2이고, 중간을 (360, 384)로 45도 돌린 이미지
axes[2].axis('off')

plt.show()
```
