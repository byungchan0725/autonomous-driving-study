```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 2, figsize=(5, 5))

dst = rgb_image[200:400, 1100:1300].copy()

axes[0].imshow(rgb_image)
axes[0].set_title('Origin Image')
axes[0].axis('off')

axes[1].imshow(dst)
axes[1].set_title('dst')
axes[1].axis('off')

plt.show()
```

```python
#      
for i in range(1100, 300, -200):
    rgb_image[200:400, i-200:i] = dst

plt.imshow(rgb_image)
plt.show()
```
