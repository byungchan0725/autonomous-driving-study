```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

dst_1 = cv2.blur(image, (10,10)) # 일반 blur
dst_2 = cv2.GaussianBlur(image, (7, 7), 0) # gaussian blur 
dst_3 = cv2.medianBlur(image, 9) # median blur
dst_4 = cv2.bilateralFilter(image, 5, 75, 75) # bilateral filterning

titles = ['Origin', 'Blur', 'Gaussuan', 'Median', 'bilateral']
images = [image, dst_1, dst_2, dst_3, dst_4]

plt.figure(figsize=(20, 6))
for i in range(len(titles)):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
