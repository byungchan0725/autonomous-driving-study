```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/gradient.jpeg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

_, th_1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
_, th_2 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
_, th_3 = cv2.threshold(image, 127, 255, cv2.THRESH_MASK)
_, th_4 = cv2.threshold(image, 127, 255, cv2.THRESH_OTSU)

titles = ['origin_rgb', 'origin_gray', 'threshold_1', 'threshold_2', 'threshold_3', 'threshold_4']
images = [rgb_image, image, th_1, th_2, th_3, th_4]

plt.figure(figsize=(20, 10))
for i in range(len(titles)):
    plt.subplot(1, len(titles), i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis('off')

plt.show()
```
