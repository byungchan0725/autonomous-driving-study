```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rgb_image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

_, th_1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

"""
- AdaptiveThresholdTypes

src  grayscale image
maxValue  
adaptiveMethod  thresholding value   
thresholdType  threshold type
blockSize  thresholding   
C     
"""

th_2 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 2)
th_3 = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 2)

titles = ['origin_rgb', 'origin_gray', 'global', 'mean', 'gaussuan']
images = [rgb_image, image, th_1, th_2, th_3]

plt.figure(figsize=(20, 20))
for i in range(len(titles)):
    plt.subplot(1, len(titles), i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis('off')

plt.show()
```
