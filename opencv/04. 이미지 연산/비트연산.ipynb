```python
import cv2 
import matplotlib.pyplot as plt

image_1 = cv2.imread('../data/bit_image_1.jpeg')
image_2 = cv2.imread('../data/bit_image_2.jpeg')

fig, axes = plt.subplots(1, 2, figsize=(10, 10))
axes[0].imshow(image_1)
axes[0].set_title('image_1')
axes[0].axis('off')

axes[1].imshow(image_2)
axes[1].set_title('image_2')
axes[1].axis('off')
plt.show()

print(image_1.shape)
print(image_2.shape)

""" 이미지에 대한 출처는 copycoding님의 티스토리를 참고하였습니다. """
```

```python
dst_1 = cv2.bitwise_and(image_1, image_2)
dst_2 = cv2.bitwise_or(image_1, image_2)
dst_3 = cv2.bitwise_not(image_1)
dst_4 = cv2.bitwise_xor(image_1, image_2)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].imshow(dst_1)
axes[0, 0].set_title('bitwise_and')
axes[0, 0].axis('off')

axes[0, 1].imshow(dst_2)
axes[0, 1].set_title('bitwise_or')
axes[0, 1].axis('off')

axes[1, 0].imshow(dst_3)
axes[1, 0].set_title('bitwise_not')
axes[1, 0].axis('off')

axes[1, 1].imshow(dst_4)
axes[1, 1].set_title('bitwise_xor')
axes[1, 1].axis('off')
plt.show()
```
