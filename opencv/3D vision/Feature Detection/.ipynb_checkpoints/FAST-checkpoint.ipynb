```python
import cv2 
import matplotlib.pyplot as plt 

image = cv2.imread('../data/checkerboard.jpeg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

fast = cv2.FastFeatureDetector_create()
keypoints_fast = fast.detect(image, None) 
# fast는 orb와 다르게 descriptors(기술자)를 넘겨주지 않는 것 같다.

image_fast = cv2.drawKeypoints(image, keypoints_fast, None, color=(0, 255, 0))

plt.imshow(image_fast)
plt.axis('off')
plt.show()
```
