```python
import cv2 
import matplotlib.pyplot as plt 

image = cv2.imread('../data/checkerboard.jpeg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

orb = cv2.ORB_create()
# ORB_create
# - nFeatures:    ( 500)
# - scaleFactor:     ( 1.2)
# - nLevels:     (  8)
# - edgeThreshold:     ( 31)
# - firstLevel:      ( 0)
# - WTA_K: Winning-Tree (K) ( 2)
# - scroreType:       ( HARRIS_SCORE)
# = pathSize:    ( 31)
# - fastThreshold: FAST     ( 20)

keypoints_orb, descriptors = orb.detectAndCompute(image, None)
# detectAndCompute
# -       
# keypoints_orb:           
# descriptors:     ()
# none  mask 

image_orb = cv2.drawKeypoints(image, keypoints_orb, None, color=(0, 255, 0))
# drawKeypoints
# - image:   
# - keypoints: keypoints  
# - output+_image:  None 
# - Color:   
# - flag:  ( ) 

plt.imshow(image_orb)
plt.axis('off')
plt.show()
```
