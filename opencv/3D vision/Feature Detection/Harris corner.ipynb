```python
import cv2 
import matplotlib.pyplot as plt 

image = cv2.imread('../data/checkerboard.jpeg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Harris corner 파라미터 설정
block_size = 2 # 코너 검출을 위한 이웃 픽셀의 크기 
ksize = 3 # Sobel 연산자 커널 크기  
k = 0.04 # Harris 코너 감도 파라미터 

corners = cv2.cornerHarris(image, block_size, ksize, k)

threshold = 0.01 
corner_mask = corners > threshold * corners.max()
# 수식을 봤을 때 λ1, λ2가 있다면 
# - λ1 >> λ2 -> Edge 
# - λ2 >> λ1 -> Edge 
# - λ1 ~ λ2 -> Corner

image[corner_mask] = 255
image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

plt.imshow(image)
plt.axis('off')
plt.show()
```
