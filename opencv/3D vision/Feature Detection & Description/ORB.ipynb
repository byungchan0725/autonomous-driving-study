```python
import cv2 
import matplotlib.pyplot as plt 

image = cv2.imread('../data/checkerboard.jpeg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

orb = cv2.ORB_create()
# ORB_create
# - nFeatures: 생성할 특징점의 개수 (기본값 500)
# - scaleFactor: 이미지 피라미드의 크기 비율 (기본값 1.2)
# - nLevels: 이미지 피라미드의 수준 수 (기본 값 8)
# - edgeThreshold: 에지 검출에 사용되는 임계값 (기본값 31)
# - firstLevel: 이미지 피라미드의 첫 번째 수준 (기본값 0)
# - WTA_K: Winning-Tree 차원(K) (기본값 2)
# - scroreType: 특징점을 선택하는데 사용되는 점수 함수의 유형 (기본값 HARRIS_SCORE)
# = pathSize: 이미지 패치의 크기 (기본값 31)
# - fastThreshold: FAST 특징점 검출에서 사용되는 임계값 (기본값 20)

keypoints_orb, descriptors = orb.detectAndCompute(image, None)
# detectAndCompute
# - 이미지에서 특징점을 찾고 기술자를 계산하는 메서드 
# keypoints_orb: 검출된 각 특징점의 좌표와 추가 정보를 포함하는 리스트 또는 배열 
# descriptors: 검출된 각 특징점에 대한 설명자료(벡터)
# none 자리에는 mask가 들어감

image_orb = cv2.drawKeypoints(image, keypoints_orb, None, color=(0, 255, 0))
# drawKeypoints
# - image: 입력 이미지 
# - keypoints: keypoints들의 리스트 
# - output+_image: 지금은 None로 설정
# - Color: 점 색 
# - flag: 플레그 (생략 가능) 

plt.imshow(image_orb)
plt.axis('off')
plt.show()
```
