```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/elon musk.jpg'
image = cv2.imread(path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

dst_1 = cv2.Canny(image, 50, 150)

dst_2 = cv2.GaussianBlur(image, (3, 3), 0) #    Canny  
dst_2 = cv2.Canny(image, 50, 150)

titles = ['origin', 'canny', 'canny with gaussian']
images = [image, dst_1, dst_2]

plt.figure(figsize=(20, 10))
for i in range(len(titles)):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
