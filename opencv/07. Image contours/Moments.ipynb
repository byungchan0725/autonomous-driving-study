```python
import cv2 
import matplotlib.pyplot as plt 

path = '../data/moments.jpg'
image = cv2.imread(path)
dst = image.copy() 

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(image, 122, 255, cv2.THRESH_BINARY_INV)

contours, hierarchy = cv2.findContours(threshold, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

M = cv2.moments(contours[0])
print(M.items)
```

```python
for i in contours:
    M = cv2.moments(i)
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])

    cv2.circle(dst, (cx, cy), 3, (255, 0, 0), -1)
    cv2.drawContours(dst, [i], 0, (0, 0, 255), 2)

plt.imshow(dst)
plt.show()
```
